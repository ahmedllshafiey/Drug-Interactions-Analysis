```python
import matplotlib.pyplot as plt

# Data
labels = ['Dentist', 'Medical Doctor', 'Nurse', 'Pharmacist', 'Physiotherapist', 'Others']
sizes = [4, 32, 9, 119, 6, 5]
```

```python
# Enhanced pie chart with explode, custom colors, and shadow
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']
explode = [0.05 if size == max(sizes) else 0 for size in sizes]  # highlight the largest slice

plt.figure(figsize=(8,8))
plt.pie(
    sizes, 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors, 
    explode=explode, 
    textprops={'fontsize': 14}
)
plt.title('Profession Distribution', fontsize=18, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Visualize key survey results with clear, focused plots

import matplotlib.pyplot as plt

# Example 1: Primary Source of Information (Top 4)
info_sources = [
    "Medical applications like Medscape & Micromedex",
    "Work colleagues and personal experiences",
    "Scientific journals",
    "Training courses and workshops"
]
info_counts = [174, 162, 155, 145]
plt.figure(figsize=(8,5))
plt.barh(info_sources[::-1], info_counts[::-1], color='#66b3ff')
plt.xlabel('Number of Respondents')
plt.title('Primary Source of Information (Top 4)')
plt.tight_layout()
plt.show()

# Example 2: Do you rely on any software or electronic applications to check for drug interactions?
app_labels = ['Yes', 'No']
app_counts = [101, 74]
plt.figure(figsize=(5,5))
plt.pie(app_counts, labels=app_labels, autopct='%1.1f%%', colors=['#99ff99', '#ff9999'], startangle=90)
plt.title('Use of Software for Drug Interaction Checks')
plt.axis('equal')
plt.show()

# Example 3: Do you document harmful drug interactions in patient records?
doc_labels = ['Yes', 'No']
doc_counts = [101, 74]
plt.figure(figsize=(5,5))
plt.pie(doc_counts, labels=doc_labels, autopct='%1.1f%%', colors=['#66b3ff', '#ffcc99'], startangle=90)
plt.title('Documentation of Harmful Drug Interactions')
plt.axis('equal')
plt.show()

# Example 4: Need for more training about drug interactions
train_labels = ['Yes, strongly', 'Yes, somewhat', 'No current training is sufficient', 'No, current training is sufficient']
train_counts = [128, 44, 2, 1]
plt.figure(figsize=(7,5))
plt.barh(train_labels[::-1], train_counts[::-1], color='#c2c2f0')
plt.xlabel('Number of Respondents')
plt.title('Need for More Training About Drug Interactions')
plt.tight_layout()
plt.show()
```
